```python
# Line Plot

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,15,7,12,8]

# Create a line plot

plt.plot(x,y,marker='+',markersize=8,color='red',linestyle='--',linewidth=2)

# add labels and titles
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("X versus Y")
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {
    'Year':[2010,2011,2012,2013,2014],
    'Sales':[500,800,1799,1500,2000]
}
df = pd.DataFrame(data)

# Create a line plot
plt.plot(df['Year'],df['Sales'],marker='o')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Over Years')
plt.grid(True)

# Set x-axis ticks at integer values
plt.xticks(df['Year'])

plt.show()
```

```python
# Bar Plot

data = {
    'Category':['A','B','C','D'],
    'Sales':[500,800,1200,1500]
}

df = pd.DataFrame(data)

# Create a bar plot

plt.bar(df['Category'],df['Sales'],color='green')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales by Category')
plt.grid(True)
```

```python
# Scatter Plot

data = {
    'Age':[25,30,22,35,28],
    'Income':[5000,60000,4500,70000,55000]
}

df = pd.DataFrame(data)

# Create a bar plot

plt.scatter(df['Age'],df['Income'],color='green')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Income by Age')
plt.grid(True)
```

```python
# Histogram

# Creating the DataFrame

data = {
    'Scores':[85,78,92,65,80,75,88,90,82,70]
}

df = pd.DataFrame(data)

# Create a histogram of the data

plt.hist(df['Scores'],width=1.5,color='green')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Distribution of Scores')
plt.grid(True)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Line Plot Example

data = sns.load_dataset("tips")

avg_tip = data.groupby("size")["tip"].mean()

plt.plot(avg_tip.index, avg_tip.values, marker='o')
plt.xlabel("Party Size")
plt.ylabel("Average Tip")
plt.title("Average Tip by Party Size")
plt.show
```

```python
# Bar Chart Example

sns.barplot(data=data, x="day", y="total_bill")
plt.title("Average Total Bill by Day")
plt.show()
```

```python
# Scatter Plot Example

plt.scatter(data["total_bill"], data["tip"])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Total Bill vs. Tip")
plt.show()
```

```python
# Histogram Example

sns.histplot(data["total_bill"], bins=20, kde=True)
plt.title("Distribution of Total Bill")
plt.show()
```

```python
# Box Plot Example

sns.boxplot(data=data, x="day", y="tip")
plt.title("Tip Distribution by Day")
plt.show()
```

```python
# Heatmap Example

corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
